```python
# ðŸŸ¢ Part A â€“ Basic Exploration
# â€¢	Load the dataset and display the first 10 rows.
import __future__
import numpy as np
import pandas as pd
# Import visualization libraries and set favourite grid style
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
# Read in the dataset
iris = pd.read_csv('Iris.csv')
iris.head(10)
iris.info()


# â€¢	Use .describe() to summarize numeric features.
iris.describe()
```

```python
# â€¢	Use .describe() to summarize numeric features.
iris.describe()
```

```python
# ðŸŸ¡ Part B â€“ Filtering & Sorting
# â€¢	Filter rows where sepal_length > 5.
filtered_sepal = iris[iris['SepalLengthCm'] > 5]
print(filtered_sepal.head())
# â€¢	Select only petal_length and species columns.
petal_species = iris[['PetalLengthCm', 'Species']]
print(petal_species.head())
# â€¢	Get all rows where species is versicolor.
versicolor_rows = iris[iris['Species'] == 'Iris-versicolor']
print(versicolor_rows.head())
# â€¢	Sort the data by sepal_width in descending order.
sorted_by_sepal_width = iris.sort_values(by='SepalWidthCm', ascending=False)
print("\nðŸ”¹ Data sorted by SepalWidthCm (descending):")
print(sorted_by_sepal_width.head())

```

```python
# ðŸŸ  Part C â€“ Grouping & Aggregation
# â€¢	Group the dataset by species and compute:
grouped = iris.groupby('Species')
#    - Mean
mean_values = grouped.mean(numeric_only=True)
print("ðŸ”¹ Mean values grouped by species:")
print(mean_values)
#    - Max
max_values = grouped.max(numeric_only=True)
print("\nðŸ”¹ Max values grouped by species:")
print(max_values)
#    - Count
count_values = grouped.count()
print("\nðŸ”¹ Count of entries grouped by species:")
print(count_values)

# â€¢	Which species has the highest average petal_length?
highest_avg_petal_length_species = mean_values['PetalLengthCm'].idxmax()
highest_avg_value = mean_values['PetalLengthCm'].max()
```

```python
# ðŸ”µ Part D â€“ Visualization

# â€¢	Plot a histogram of sepal_length.
# plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x='SepalLengthCm', bins=20,kde=True, color='skyblue')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
# â€¢	Create a boxplot of petal_width by species.
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x='Species', y='PetalWidthCm', palette='Set2')
plt.title('Boxplot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.show()

# â€¢	Draw a scatter plot of sepal_length vs petal_length, color by species.
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x='SepalLengthCm', y='PetalLengthCm', hue='Species', palette='Set1')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()

# â€¢	Use seaborn.pairplot() to visualize all numerical relationships.
sns.pairplot(iris.drop(columns='Id'), hue='Species', palette='husl')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()

```
